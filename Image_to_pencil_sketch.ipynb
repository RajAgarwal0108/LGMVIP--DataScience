# Image to Pencil Sketch
## Tasks:
    1. Load the colured image
    2. Turn the colour image into gray image
    3. Invert the gray image
    4. Blur the image
    5. Invert the blur image
    6. divide the gray image with inverted blur image

```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

### Load The Image

```python
img = cv2.imread('/home/raj/Pictures/dog.jpg')
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(img)
plt.axis('off')
plt.title('Original-Image')
```

### Turn the image into gray image

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(img_gray,cmap='gray')
plt.axis('off')
plt.title('Gray-Image')
```

### Invert the gray image

```python
inverted_gray_image = 255-img_gray
plt.figure(figsize=(8,8))
plt.imshow(inverted_gray_image,cmap='gray')
plt.axis('off')
```

### Blur the inverted gray image

```python
blured_img = cv2.GaussianBlur(inverted_gray_image,(21,21),0)
plt.figure(figsize=(8,8))
plt.imshow(blured_img)
plt.axis('off')
```

### Invert the blured image

```python
inverted_blur_image = 255-blured_img
plt.figure(figsize=(8,8))
plt.imshow(inverted_blur_image)
plt.axis('off')
```

### divide the gray image with inverted blur image

```python
sketch = cv2.divide(img_gray,inverted_blur_image,scale=255)
plt.figure(figsize=(15,15))
plt.imshow(sketch,cmap='gray')
plt.axis('off')
plt.title('sketch-Image')
```
